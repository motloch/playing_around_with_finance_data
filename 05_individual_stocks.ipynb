```python
# Collect statistics about individual stocks
means = []
stds  = []

for ticker in s_and_p:
    data = pd.read_csv('individual_stocks_5yr/' + ticker + '_data.csv', parse_dates = ['date'], index_col = 0)
    data['Pct_change'] = data['close'].pct_change()        # p2/p1 - 1
    data['Log_change'] = np.log(1 + data['Pct_change'])    # Log(p2/p1)

    means.append(data['Log_change'].mean())
    stds.append(data['Log_change'].std())
```

```python
# Combine data into a dataframe
df = pd.DataFrame({'name': s_and_p, 'daily_log_gain': means, 'daily_log_std': stds})
df['annual_change'] = np.exp(df['daily_log_gain'])**250
df.head()
```

```python
# Plot the graph, with mouse hover capability - see which point is which security
fig = px.scatter(df, x="annual_change", y="daily_log_std", text='name')
fig.update_traces(hoverinfo='text', hovertemplate=None , mode='markers')
fig.update_layout(hoverdistance=7)
```
